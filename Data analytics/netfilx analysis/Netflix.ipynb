```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="whitegrid")
```

```python
df = pd.read_csv("netflix_titles.csv")
df.head(10)
```

```python

df.info()
print("\nDiscription of Dataset\n")
df.describe()
```

```python
print("Rows:", df.shape[0])
print("Columns:", df.shape[1])
```

```python
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
print("Duplicate values")
df.duplicated().sum()
```

```python
#"Filling Missing Values\n"
df['country'].fillna("Unknown", inplace=True)
df['cast'].fillna("Unknown", inplace=True)
```

```python
plt.figure(figsize=(5,4))
sns.countplot(data=df, x='type', palette='Set2')
plt.title('Movies vs TV Shows on Netflix')
plt.xlabel('Type')
plt.ylabel('Count')
```

```python
director_counts = (
    df.loc[df['director'] != 'Unknown', 'director']  # filter first
      .str.split(', ')
      .explode()
      .value_counts()
)

top_directors = director_counts.head(10)

plt.figure()
sns.barplot(x=top_directors.index,y=top_directors.values,palette="coolwarm")
plt.xlabel("Director")
plt.ylabel("Number of Titles")
plt.title("Top Directors by Number of Titles")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
df['release_year'].value_counts().sort_index().plot(kind='line')
plt.title('Number of Titles Released Each Year')
plt.xlabel('Release Year')
plt.ylabel('Count')
plt.show()
```

```python
top_countries = df['country'].value_counts().head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=top_countries.values, y=top_countries.index, palette='coolwarm')
plt.title('Top 10 Countries with Most Netflix Titles')
plt.xlabel('Number of Titles')
plt.ylabel('Country')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(data=df, y='rating', order=df['rating'].value_counts().index, palette='pastel')
plt.title('Content Ratings Distribution')
plt.xlabel('Count')
plt.ylabel('Rating')
plt.show()
```

```python
# Apply conversion to TV Shows
df_tv = df[df['type'] == 'TV Show'].copy()

df_tv['seasons'] = df_tv['duration'].apply(
    lambda d: int(d.split()[0])
)
```

```python
df_tv_genre = df_tv.assign(
    listed_in=df_tv['listed_in'].str.split(', ')
).explode('listed_in')
avg_seasons_genre = (
    df_tv_genre
    .groupby('listed_in')['seasons']
    .mean()
    .sort_values(ascending=False)
)
```

```python
plt.figure(figsize=(10,6))

sns.barplot(
    x=avg_seasons_genre.values,
    y=avg_seasons_genre.index,
    palette=['b','g','r','y']
)

plt.title('Average Number of Seasons by TV Show Genre')
plt.xlabel('Average Seasons')
plt.ylabel('Genre')
plt.show()
```
